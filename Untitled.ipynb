```python
def calculate_correlation_matrices(
    df_period_data: pd.DataFrame,
    has_sectors: bool, # Explicitly tell the function if sectors are expected
    min_trading_days: int = 0 # NEW: Minimum non-NaN data points required
) -> Dict[str, pd.DataFrame]:
    """
    Calculates correlation matrices. If has_sectors is True, expects
    MultiIndex columns (Sector, Symbol) and returns a dict of sector matrices.
    Otherwise, expects simple Symbol columns and returns a dict with one entry.

    NEW: It now filters stocks to include only those with at least
    'min_trading_days' non-NaN data points in the period.

    Diagonals are set to NaN.
    """
    sector_corr_matrices = {}

    if has_sectors:
        if not isinstance(df_period_data.columns, pd.MultiIndex) or df_period_data.columns.nlevels != 2:
            raise ValueError("Input DataFrame must have a 2-level MultiIndex (Sector, Symbol) when has_sectors=True.")

        df_transposed = df_period_data.T
        # Group by 'Sector' (level 0 of the new index)
        grouped_by_sector_T = df_transposed.groupby(level=0)

        for sector_name, group_df_T in grouped_by_sector_T:
            sector_df = group_df_T.T

            # --- START NEW FILTERING BLOCK ---
            # 1. Count non-NaN days *before* dropping level
            counts = sector_df.count() # Counts non-NaNs per (Sector, Symbol) column
            # 2. Identify columns (MultiIndex tuples) that meet the threshold
            valid_multi_cols = counts[counts >= min_trading_days].index
            # 3. Filter the DataFrame *before* dropping level
            sector_df_filtered = sector_df[valid_multi_cols].copy()
            # --- END NEW FILTERING BLOCK ---

            # Check if enough stocks remain after filtering
            if sector_df_filtered.shape[1] < 2:
                # print(f"Warning: Sector '{sector_name}' has < 2 stocks after filtering. Skipping.")
                continue

            # Ensure columns are just symbols for .corr()
            sector_df_filtered.columns = sector_df_filtered.columns.get_level_values(1) # Get 'Symbol' level
            
            # Remove potential duplicate columns after droplevel
            sector_df_filtered = sector_df_filtered.loc[:, ~sector_df_filtered.columns.duplicated()]

            correlation_matrix = sector_df_filtered.corr()
            np.fill_diagonal(correlation_matrix.values, np.nan)
            sector_corr_matrices[sector_name] = correlation_matrix
            
    else: # No sectors, calculate one correlation matrix for all assets
        df_to_process = df_period_data.copy()
        
        # Ensure columns are simple if starting with MultiIndex
        if isinstance(df_to_process.columns, pd.MultiIndex):
            print("Warning: has_sectors=False but DataFrame has MultiIndex columns. Using level 1 as symbols.")
            df_to_process.columns = df_to_process.columns.get_level_values(1)

        # --- START NEW FILTERING BLOCK ---
        counts = df_to_process.count()
        valid_cols = counts[counts >= min_trading_days].index
        df_filtered = df_to_process[valid_cols].copy()
        # --- END NEW FILTERING BLOCK ---

        # Check if enough stocks remain after filtering
        if df_filtered.shape[1] < 2:
            print("Warning: 'all_assets' has < 2 stocks after filtering. Returning empty dict.")
            return {}

        correlation_matrix = df_filtered.corr()
        np.fill_diagonal(correlation_matrix.values, np.nan)
        sector_corr_matrices['all_assets'] = correlation_matrix # Use a default key

    return sector_corr_matrices
```

```python
import pandas as pd
import numpy as np
from typing import Dict, Optional, List, Tuple, Union
def get_top_pairs_from_matrices(
    sector_corr_matrices: Dict[str, pd.DataFrame],
    quantity: int,
    min_corr_threshold: Optional[float] = None,
    max_correlation_threshold: Optional[float] = None,
    include_correlation_data: bool = True  # New parameter, defaults to current behavior
) -> Dict[str, Union[pd.DataFrame, List[Tuple[str, str]]]]: # Updated return type
    """
    Extracts top N correlated pairs for each sector, with optional thresholds.
    Can return either a DataFrame with correlation data or a list of stock pair tuples.

    Args:
        sector_corr_matrices: A dictionary where keys are sector names (str)
                              and values are pandas DataFrames representing
                              the correlation matrix for that sector (ideally
                              with NaN on the diagonal).
        quantity: The number of top correlated pairs to return for each sector
                  after filtering.
        min_corr_threshold: If provided, only pairs with a correlation
                            strictly greater than this value will be considered.
        max_correlation_threshold: If provided, only pairs with a correlation
                                   strictly less than this value will be considered.
        include_correlation_data: If True (default), returns a DataFrame including
                                  Stock1, Stock2, and Correlation.
                                  If False, returns a list of tuples (Stock1, Stock2).

    Returns:
        A dictionary where keys are sector names and values are either DataFrames
        or lists of tuples, based on 'include_correlation_data'.
    """
    top_pairs_result = {}
    for sector, mat in sector_corr_matrices.items():
        # print(f"--- Processing: {sector} ---") # Optional: for debugging

        if mat.shape[0] <= 1: # Need at least 2 stocks to form a pair
            if include_correlation_data:
                top_pairs_result[sector] = pd.DataFrame(columns=['Stock1', 'Stock2', 'Correlation'])
            else:
                top_pairs_result[sector] = []
            continue

        try:
            stacked_series = mat.stack(future_stack=True)
            if not isinstance(stacked_series, pd.Series):
                if isinstance(stacked_series, pd.DataFrame) and stacked_series.shape[0] == 1:
                    stacked_series = stacked_series.iloc[0] # Try to get the Series from 1-row DF
                else:
                    print(f"[ERROR] .stack() for {sector} did not produce a Series. Got {type(stacked_series)}. Skipping.")
                    if include_correlation_data:
                        top_pairs_result[sector] = pd.DataFrame(columns=['Stock1', 'Stock2', 'Correlation'])
                    else:
                        top_pairs_result[sector] = []
                    continue
            
            sorted_pairs = stacked_series.sort_values(ascending=False)
            pairs_df = sorted_pairs.reset_index()
            pairs_df.columns = ['Stock1', 'Stock2', 'Correlation']
            pairs_df = pairs_df.dropna()
            pairs_df = pairs_df[pairs_df['Stock1'] != pairs_df['Stock2']].copy()
            pairs_df['pair_key'] = pairs_df.apply(lambda row: frozenset([row['Stock1'], row['Stock2']]), axis=1)
            unique_pairs = pairs_df.drop_duplicates(subset='pair_key')

            if min_corr_threshold is not None:
                unique_pairs = unique_pairs[unique_pairs['Correlation'] > min_corr_threshold]
            
            if max_correlation_threshold is not None:
                unique_pairs = unique_pairs[unique_pairs['Correlation'] < max_correlation_threshold]

            # --- MODIFIED OUTPUT LOGIC ---
            if include_correlation_data:
                top_n_output = unique_pairs[['Stock1', 'Stock2', 'Correlation']].head(quantity)
            else:
                top_n_df_for_tuples = unique_pairs[['Stock1', 'Stock2']].head(quantity)
                top_n_output = [tuple(x) for x in top_n_df_for_tuples.to_numpy()]
            # --- END MODIFIED OUTPUT LOGIC ---

            top_pairs_result[sector] = top_n_output

        except Exception as e:
            print(f"[ERROR] Could not process pairs for {sector}: {e}")
            if include_correlation_data:
                top_pairs_result[sector] = pd.DataFrame(columns=['Stock1', 'Stock2', 'Correlation'])
            else:
                top_pairs_result[sector] = []
    return top_pairs_result
```

```python
def get_random_chunk(df, chunk_size):
    """
    Selects a random consecutive chunk from a Pandas DataFrame.

    Args:
      df: The Pandas DataFrame containing the time series data.
      chunk_size: The desired size of the random chunk.

    Returns:
      A Pandas DataFrame representing the random chunk, or None if the chunk size 
      is invalid.
  """
    if chunk_size <= 0 or chunk_size > len(df):
        return None

    start_index = random.randint(0, len(df) - chunk_size)
    end_index = start_index + chunk_size
    return df.iloc[start_index:end_index].copy()
```

```python
def get_data_splits(
    df: pd.DataFrame,
    in_sample_years: int,
    oos_years: int
) -> tuple[pd.DataFrame, pd.DataFrame]:
    """
    Selects a random consecutive chunk from a DataFrame based on years
    and splits it into in-sample and out-of-sample periods.

    Args:
        df: The Pandas DataFrame with a DateTime index.
        in_sample_years: The desired duration of the in-sample period in years.
        oos_years: The desired duration of the out-of-sample period in years.

    Returns:
        A tuple containing two Pandas DataFrames:
        (in_sample_df, out_of_sample_df), or (None, None) if
        the duration is invalid.
    """
    if not isinstance(df.index, pd.DatetimeIndex):
        raise TypeError("DataFrame index must be a DatetimeIndex.")

    total_years = in_sample_years + oos_years
    if total_years <= 0:
        print("Error: Total years must be positive.")
        return None, None

    # Calculate total duration in days (approximate for finding end date)
    total_duration_days = total_years * 365.25
    min_date = df.index.min()
    max_date = df.index.max()

    # Ensure we don't pick a start date too close to the end
    latest_possible_start = max_date - timedelta(days=total_duration_days)

    if latest_possible_start < min_date:
        print("Error: Requested duration is longer than the DataFrame's span.")
        return None, None

    # Pick a random start date (as an integer index)
    possible_start_indices = df.index[df.index <= latest_possible_start].size
    if possible_start_indices == 0:
        print("Error: Could not find a suitable start index.")
        return None, None
        
    start_idx = random.randint(0, possible_start_indices - 1)
    start_date = df.index[start_idx]

    # Calculate split and end dates
    split_date = start_date + pd.DateOffset(years=in_sample_years)
    end_date = split_date + pd.DateOffset(years=oos_years)

    # Find the closest actual index positions
    start_loc = df.index.get_indexer([start_date], method='nearest')[0]
    split_loc = df.index.get_indexer([split_date], method='nearest')[0]
    end_loc = df.index.get_indexer([end_date], method='nearest')[0]

    # Ensure end_loc doesn't exceed bounds
    end_loc = min(end_loc, len(df) -1)
    
    # Ensure split_loc doesn't exceed end_loc
    split_loc = min(split_loc, end_loc)

    # Slice the DataFrame
    in_sample_df = df.iloc[start_loc:split_loc].copy()
    out_of_sample_df = df.iloc[split_loc:end_loc].copy()

    return in_sample_df, out_of_sample_df
```

```python
import itertools

def find_common_pairs_from_lists(list_of_tuples1: List[Tuple[str, str]], list_of_tuples2: List[Tuple[str, str]]) -> List[Tuple[str, str]]:
    """
    Finds common pairs between two lists of (Stock1, Stock2) tuples,
    ignoring order within tuples.
    """
    set1 = {frozenset(pair) for pair in list_of_tuples1}
    set2 = {frozenset(pair) for pair in list_of_tuples2}
    
    common_frozensets = set1.intersection(set2)
    
    # Return as sorted tuples for consistency
    common_pairs_list = [tuple(sorted(list(fs_pair))) for fs_pair in common_frozensets]
    return common_pairs_list
```

```python
def test_engle_granger(
    df: pd.DataFrame,
    stock1: str,
    stock2: str
) -> Optional[Tuple[float, float, float, float]]: # Updated return type
    """
    Performs the Engle-Granger two-step cointegration test and returns parameters.

    Args:
        df: DataFrame containing the price series.
        stock1: Ticker symbol of the first stock (will be used as 'x').
        stock2: Ticker symbol of the second stock (will be used as 'y').

    Returns:
        A tuple containing (p_value, beta, spread_mean, spread_std_dev),
        or None if the test cannot be performed.
    """
    # 1. Prepare data and perform OLS
    ols_df = df[[stock1, stock2]].copy()
    ols_df.dropna(inplace=True)

    y = ols_df[stock2]
    x_val = ols_df[stock1] # Renamed to avoid confusion
    x = sm.add_constant(x_val) # Add constant for OLS

    result = sm.OLS(y, x).fit()
    residuals = result.resid

    # **NEW**: Extract Beta (hedge ratio)
    # result.params[0] is the constant, result.params[1] is the beta for stock1
    beta = result.params[stock1]

    # **NEW**: Calculate spread mean and std dev
    spread_mean = residuals.mean()
    spread_std_dev = residuals.std()

    # 2. Perform ADF test on residuals
    adf_result = sm.tsa.stattools.adfuller(residuals, regression='c', autolag='AIC')
    p_value = adf_result[1]

    # **NEW**: Return all values
    return p_value, beta, spread_mean, spread_std_dev #
```

```python
def filter_cointegrated_pairs(
    df: pd.DataFrame,
    pairs_list: List[Tuple[str, str]],
    p_value_threshold: float = 0.05
) -> Dict[Tuple[str, str], Dict[str, float]]: # Updated return type
    """
    Filters stock pairs for cointegration and stores their parameters.

    Args:
        df: DataFrame containing the price series for the testing period. [cite: 26]
        pairs_list: A list of tuples, where each tuple is a stock pair. [cite: 27]
        p_value_threshold: The significance level for the ADF test. [cite: 27]

    Returns:
        A dictionary where keys are (Stock1, Stock2) tuples and values are
        dictionaries containing 'p_value', 'beta', 'mean', and 'std_dev'.
    """
    cointegrated_results = {} # Changed name and structure
    print(f"Testing {len(pairs_list)} pairs for cointegration...") # [cite: 27]

    for stock1, stock2 in pairs_list: # [cite: 27]
        # Ensure we don't test a pair if its reverse is already found
        if (stock1, stock2) not in cointegrated_results and \
           (stock2, stock1) not in cointegrated_results:

            test_result = test_engle_granger(df, stock1, stock2) # Call modified function

            if test_result is not None:
                p_value, beta, spread_mean, spread_std_dev = test_result # Unpack

                if p_value <= p_value_threshold:
                    # Store results with pair as key and params as value
                    cointegrated_results[(stock1, stock2)] = {
                        'p_value': p_value,
                        'beta': beta,
                        'mean': spread_mean,
                        'std_dev': spread_std_dev
                    }
                    # print(f"Found: ({stock1}, {stock2}) p={p_value:.3f}, b={beta:.2f}")

    print(f"Found {len(cointegrated_results)} cointegrated pairs.") # [cite: 28]
    return cointegrated_results
```

```python
def map_bands(spread, bands):

    if pd.isna(spread):
        return 0 # Or return np.nan if you prefer
    
    if spread < bands[1]:
        spread = 1
    elif spread > bands[0]:
        spread = -1
    else:
        spread = 0
    return spread
```

```python
spread_returns_df = pd.DataFrame(index=daily_returns_df.index)

for pair, params in cointegrated_results.items():
    stock1, stock2 = pair
    beta = params['beta']

    if stock1 in daily_returns_df.columns and stock2 in daily_returns_df.columns:
        spread_returns_df[pair] = daily_returns_df[stock2] - beta * daily_returns_df[stock1]

initial_capital = 100000
current_capital = initial_capital
equity_curve_values = [initial_capital] # To store daily equity

# DataFrames to store the number of shares held for each stock in each pair
# Initialize with zeros, same index and columns as positions_df
shares_S1_df = pd.DataFrame(0.0, index=positions_df.index, columns=positions_df.columns)
shares_S2_df = pd.DataFrame(0.0, index=positions_df.index, columns=positions_df.columns)

# DataFrame to store daily P&L for each pair
daily_pair_pnl_df = pd.DataFrame(0.0, index=positions_df.index, columns=positions_df.columns)

for i in range(1, len(out_of_sample_df.index)): # Start from the second day
    today = out_of_sample_df.index[i]
    yesterday = out_of_sample_df.index[i-1]
    
    # Update current capital based on *yesterday's* P&L for today's allocation
    # (or use capital at start of day before today's trades for simplicity)
    if i > 1: # After the first P&L day
        current_capital += total_daily_pnl_for_yesterday # We'll calculate this

    capital_per_trade = current_capital * 0.05
    capital_per_leg = capital_per_trade / 2.0

    today_total_pnl = 0.0

    for pair in positions_df.columns: # Iterate through all potential pairs
        stock1, stock2 = pair
        # beta = cointegrated_results[pair]['beta'] # Beta is for signal, not this sizing method

        current_position_signal = positions_df.loc[today, pair]
        yesterday_holding = holdings_df.loc[today, pair] # Position held coming into today

        # 1. Handle Entries
        if yesterday_holding == 0 and current_position_signal != 0: # New trade entry
            price_s1_at_entry = out_of_sample_df.loc[yesterday, stock1] # Price when decision was made
            price_s2_at_entry = out_of_sample_df.loc[yesterday, stock2]

            if pd.notna(price_s1_at_entry) and price_s1_at_entry > 0 and \
               pd.notna(price_s2_at_entry) and price_s2_at_entry > 0:

                if current_position_signal == 1: # Long the spread (Long S2, Short S1)
                    shares_S2_df.loc[today, pair] = capital_per_leg / price_s2_at_entry
                    shares_S1_df.loc[today, pair] = - (capital_per_leg / price_s1_at_entry)
                elif current_position_signal == -1: # Short the spread (Short S2, Long S1)
                    shares_S2_df.loc[today, pair] = - (capital_per_leg / price_s2_at_entry)
                    shares_S1_df.loc[today, pair] = capital_per_leg / price_s1_at_entry
            else: # Cannot enter if price is NaN or zero
                 shares_S2_df.loc[today, pair] = 0
                 shares_S1_df.loc[today, pair] = 0


        # 2. Handle Exits
        elif yesterday_holding != 0 and current_position_signal == 0: # Exiting a trade
            # Shares go to zero, P&L for the day will be calculated below
            shares_S2_df.loc[today, pair] = 0
            shares_S1_df.loc[today, pair] = 0
            # Note: Real P&L for exit includes today's price move up to exit point.

        # 3. Maintain Holdings
        elif yesterday_holding != 0 and current_position_signal == yesterday_holding:
            # Position is maintained, copy shares from yesterday
            shares_S2_df.loc[today, pair] = shares_S2_df.loc[yesterday, pair]
            shares_S1_df.loc[today, pair] = shares_S1_df.loc[yesterday, pair]
        
        # 4. Handle Flips (e.g., Long to Short) - simplified as exit then entry
        elif yesterday_holding != 0 and current_position_signal == -yesterday_holding:
            # First, treat as an exit (shares go to 0 for today's P&L calc on old position)
            # Then, treat as a new entry for the *next* day's holding (already handled by #1 on next iter)
            # For P&L calculation for *today*, we use *yesterday's* shares.
            # For the *new* position, it will be sized based on *today's* prices to start *tomorrow*.
            # So for today, we exit. The new entry will be sized and reflected in shares_df tomorrow.
             shares_S2_df.loc[today, pair] = 0 # Mark as exited for today's P&L calc
             shares_S1_df.loc[today, pair] = 0

        # 5. Calculate P&L for *today* based on shares held *coming into today*
        if yesterday_holding != 0: # If there was a position coming into today
            price_s1_yesterday = out_of_sample_df.loc[yesterday, stock1]
            price_s2_yesterday = out_of_sample_df.loc[yesterday, stock2]
            price_s1_today = out_of_sample_df.loc[today, stock1]
            price_s2_today = out_of_sample_df.loc[today, stock2]

            # Ensure all prices are valid
            if pd.notna(price_s1_yesterday) and pd.notna(price_s2_yesterday) and \
               pd.notna(price_s1_today) and pd.notna(price_s2_today):
                
                pnl_s1 = shares_S1_df.loc[yesterday, pair] * (price_s1_today - price_s1_yesterday)
                pnl_s2 = shares_S2_df.loc[yesterday, pair] * (price_s2_today - price_s2_yesterday)
                pair_pnl = pnl_s1 + pnl_s2
                daily_pair_pnl_df.loc[today, pair] = pair_pnl
                today_total_pnl += pair_pnl
            else: # If price is NaN, P&L for this pair is 0 for the day
                daily_pair_pnl_df.loc[today, pair] = 0
    
    # Update equity curve
    current_capital += today_total_pnl 
    equity_curve_values.append(current_capital)
    total_daily_pnl_for_yesterday = today_total_pnl # For next day's capital allocation

# Convert equity_curve_values to a Pandas Series
equity_curve = pd.Series(equity_curve_values, index=out_of_sample_df.index)
equity_curve.plot(title="Strategy Equity Curve")
plt.xlabel("DateTime") # Optional: Add x-axis label
plt.ylabel("Portfolio Equity") # Optional: Add y-axis label
plt.show() # Ensures the plot is displayed
```

```python
# --- User-Defined Cost Parameters ---
COMMISSION_PER_SHARE = 0.005  # Example: $0.005 per share
MIN_COMMISSION_PER_ORDER_LEG = 1.00 # Example: Minimum $1 per leg of the trade
MAX_COMMISSION_PER_ORDER_LEG_PCT = 0.01 # Example: Max 1% of trade value for a leg

SLIPPAGE_PER_SHARE = 0.01 # Example: $0.01 slippage per share transacted

ANNUAL_BORROW_RATE_ESTIMATE = 0.01
DAILY_BORROW_RATE = ANNUAL_BORROW_RATE_ESTIMATE / 252.0

# --- Dividend Schedule (Example: you'll need to populate this) ---
# Format: pd.DataFrame with MultiIndex (date, ticker) and 'dividend_amount' column
# Or a dictionary: {(date, ticker): amount}
# For simplicity, we'll assume a function `get_dividend(date, ticker)` exists
# Example: dividend_schedule = pd.read_csv('your_dividend_data.csv').set_index(['date', 'ticker'])
def get_dividend(date, ticker, dividend_schedule_df):
    try:
        return dividend_schedule_df.loc[(date, ticker), 'dividend_amount']
    except KeyError:
        return 0.0

# --- Initialization ---
current_capital = initial_capital
equity_curve_values = [initial_capital]
daily_pnl_values = [0.0] # To store net daily P&L

# DataFrames to store shares. Initialize with previous day's index for correct alignment in loop
# Ensure consistent indexing from the start of the OOS period
shares_S1_df = pd.DataFrame(0.0, index=out_of_sample_df.index, columns=positions_df.columns)
shares_S2_df = pd.DataFrame(0.0, index=out_of_sample_df.index, columns=positions_df.columns)

# --- Daily Backtesting Loop ---
for i in range(1, len(out_of_sample_df.index)):
    today = out_of_sample_df.index[i]
    yesterday = out_of_sample_df.index[i-1]

    # Capital for new trades is based on capital at the START of 'today' (i.e., end of 'yesterday')
    capital_for_new_trades = equity_curve_values[-1] # Current capital before today's P&L/costs
    capital_per_trade = capital_for_new_trades * 0.05
    capital_per_leg = capital_per_trade / 2.0

    today_total_gross_pnl = 0.0  # P&L from price movements
    today_total_commissions = 0.0
    today_total_slippage_cost = 0.0
    today_total_borrow_costs = 0.0
    today_total_dividend_costs = 0.0

    for pair_tuple in positions_df.columns: # Renamed 'pair' to 'pair_tuple' to avoid conflict
        stock1, stock2 = pair_tuple

        # Prices needed for calculations
        price_s1_yesterday = out_of_sample_df.loc[yesterday, stock1]
        price_s2_yesterday = out_of_sample_df.loc[yesterday, stock2]
        price_s1_today = out_of_sample_df.loc[today, stock1]
        price_s2_today = out_of_sample_df.loc[today, stock2]

        # Signals and holdings
        current_signal = positions_df.loc[today, pair_tuple]
        shares_s1_held_yesterday = shares_S1_df.loc[yesterday, pair_tuple]
        shares_s2_held_yesterday = shares_S2_df.loc[yesterday, pair_tuple]

        new_shares_s1 = shares_s1_held_yesterday
        new_shares_s2 = shares_s2_held_yesterday
        
        traded_shares_s1 = 0
        traded_shares_s2 = 0

        # --- 1. Determine Trades & Calculate Shares for New/Changed Positions ---
        # Entering a new position or flipping
        if (shares_s1_held_yesterday == 0 and current_signal != 0) or \
           (shares_s1_held_yesterday != 0 and current_signal == -np.sign(shares_s1_held_yesterday)): # Flip

            # If flipping, conceptually close old position (shares go to 0 for cost calc)
            # and then open new one. Costs apply to both parts of the flip.
            if shares_s1_held_yesterday != 0: # It's a flip, so account for closing the old position
                traded_shares_s1 += abs(shares_s1_held_yesterday)
                traded_shares_s2 += abs(shares_s2_held_yesterday)

            if pd.notna(price_s1_yesterday) and price_s1_yesterday > 0 and \
               pd.notna(price_s2_yesterday) and price_s2_yesterday > 0:
                
                if current_signal == 1: # Long Spread (Long S2, Short S1)
                    new_shares_s2 = capital_per_leg / price_s2_yesterday
                    new_shares_s1 = - (capital_per_leg / price_s1_yesterday)
                elif current_signal == -1: # Short Spread (Short S2, Long S1)
                    new_shares_s2 = - (capital_per_leg / price_s2_yesterday)
                    new_shares_s1 = capital_per_leg / price_s1_yesterday
                
                traded_shares_s1 += abs(new_shares_s1)
                traded_shares_s2 += abs(new_shares_s2)
            else: # Cannot size if price is invalid
                new_shares_s1 = 0
                new_shares_s2 = 0
        
        # Exiting a position
        elif shares_s1_held_yesterday != 0 and current_signal == 0:
            traded_shares_s1 = abs(shares_s1_held_yesterday)
            traded_shares_s2 = abs(shares_s2_held_yesterday)
            new_shares_s1 = 0
            new_shares_s2 = 0

        shares_S1_df.loc[today, pair_tuple] = new_shares_s1
        shares_S2_df.loc[today, pair_tuple] = new_shares_s2

        # --- 2. Calculate Commissions and Slippage for today's trades ---
        if traded_shares_s1 > 0:
            # S1 leg commission
            comm_s1 = traded_shares_s1 * COMMISSION_PER_SHARE
            comm_s1 = max(comm_s1, MIN_COMMISSION_PER_ORDER_LEG if traded_shares_s1 > 0 else 0) # Min commission
            comm_s1 = min(comm_s1, abs(traded_shares_s1 * price_s1_yesterday * MAX_COMMISSION_PER_ORDER_LEG_PCT) if pd.notna(price_s1_yesterday) and price_s1_yesterday > 0 else comm_s1 ) # Max commission
            today_total_commissions += comm_s1
            today_total_slippage_cost += traded_shares_s1 * SLIPPAGE_PER_SHARE

        if traded_shares_s2 > 0:
            # S2 leg commission
            comm_s2 = traded_shares_s2 * COMMISSION_PER_SHARE
            comm_s2 = max(comm_s2, MIN_COMMISSION_PER_ORDER_LEG if traded_shares_s2 > 0 else 0)
            comm_s2 = min(comm_s2, abs(traded_shares_s2 * price_s2_yesterday * MAX_COMMISSION_PER_ORDER_LEG_PCT) if pd.notna(price_s2_yesterday) and price_s2_yesterday > 0 else comm_s2)
            today_total_commissions += comm_s2
            today_total_slippage_cost += traded_shares_s2 * SLIPPAGE_PER_SHARE

        # --- 3. Calculate Gross P&L from price movements (based on shares held *from yesterday*) ---
        if shares_s1_held_yesterday != 0 or shares_s2_held_yesterday != 0: # If a position was held
            if pd.notna(price_s1_today) and pd.notna(price_s1_yesterday):
                pnl_s1 = shares_s1_held_yesterday * (price_s1_today - price_s1_yesterday)
                today_total_gross_pnl += pnl_s1
            if pd.notna(price_s2_today) and pd.notna(price_s2_yesterday):
                pnl_s2 = shares_s2_held_yesterday * (price_s2_today - price_s2_yesterday)
                today_total_gross_pnl += pnl_s2
        
        # --- 4. Calculate Daily Stock Borrow Costs & Dividend Adjustments (for positions held *overnight yesterday*) ---
        # S1 was short
        if shares_s1_held_yesterday < 0 and pd.notna(price_s1_yesterday):
            market_value_s1_short = abs(shares_s1_held_yesterday) * price_s1_yesterday
            today_total_borrow_costs += market_value_s1_short * DAILY_BORROW_RATE
            dividend_s1 = get_dividend(today, stock1, dividend_schedule_df) # Assumes dividend_schedule_df exists
            if dividend_s1 > 0:
                today_total_dividend_costs += abs(shares_s1_held_yesterday) * dividend_s1
        
        # S2 was short
        if shares_s2_held_yesterday < 0 and pd.notna(price_s2_yesterday):
            market_value_s2_short = abs(shares_s2_held_yesterday) * price_s2_yesterday
            today_total_borrow_costs += market_value_s2_short * DAILY_BORROW_RATE
            dividend_s2 = get_dividend(today, stock2, dividend_schedule_df) # Assumes dividend_schedule_df exists
            if dividend_s2 > 0:
                today_total_dividend_costs += abs(shares_s2_held_yesterday) * dividend_s2
                
    # --- 5. Calculate Net P&L for the day and update capital ---
    net_daily_pnl = today_total_gross_pnl - today_total_commissions - today_total_slippage_cost - today_total_borrow_costs - today_total_dividend_costs
    daily_pnl_values.append(net_daily_pnl) # Store for analysis
    
    # Update capital using the capital *before* today's P&L and costs for allocation consistency
    current_capital = equity_curve_values[-1] + net_daily_pnl
    equity_curve_values.append(current_capital)

# --- Finalize Equity Curve ---
equity_curve = pd.Series(equity_curve_values[1:], index=out_of_sample_df.index[1:]) # Match length, skip initial capital

# Plot
equity_curve.plot(title="Strategy Equity Curve (with Costs)")
plt.xlabel("DateTime")
plt.ylabel("Portfolio Equity")
plt.show()

# You can also analyze daily_pnl_values
daily_pnl_series = pd.Series(daily_pnl_values[1:], index=out_of_sample_df.index[1:])
```
